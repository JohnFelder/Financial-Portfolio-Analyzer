```python
import os
import json
import requests
import pandas as pd
from dotenv import load_dotenv
```

```python
# Load env environment variables
load_dotenv()

# Set crypto_api_key
crypto_api_key = os.getenv('crypto_api_key')

# Set the data endpoint
request_url = 'https://min-api.cryptocompare.com/data/pricemultifull?fsyms=BTC,ATOM,MATIC,UNI,CAKE&tsyms=USD&api_key='

# Concatenate endpoint and key
crypto_request_url = request_url + crypto_api_key

# Fetch the data, stored as json
response = requests.get(crypto_request_url).json()

# Displays the response data
print(json.dumps(response, indent=4, sort_keys=False))
```
